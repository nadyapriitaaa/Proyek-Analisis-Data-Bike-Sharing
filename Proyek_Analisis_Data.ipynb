### Cleaning Data

```python
hour_df = hour_df.drop("instant", axis=1)
```

```python
day_df = day_df.drop("instant", axis=1)
```

```python
hour_df.season = hour_df.season.cat.rename_categories({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
hour_df.mnth = hour_df.mnth.cat.rename_categories({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})
hour_df.weekday = hour_df.weekday.replace({0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'})
hour_df.workingday = hour_df.workingday.cat.rename_categories({0: 'No', 1: 'Yes'})
hour_df.yr = hour_df.yr.replace({0: 2011, 1: 2012})
hour_df.hr = hour_df.hr.apply(lambda x: str(x) + ' AM' if x < 12 else str(x-12) + ' PM' if x != 24 else '12 PM')
```

```python
day_df.season = day_df.season.cat.rename_categories({1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'})
day_df.mnth = day_df.mnth.cat.rename_categories({1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June', 7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'})
day_df.weekday = day_df.weekday.replace({0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday'})
day_df.workingday = day_df.workingday.cat.rename_categories({0: 'No', 1: 'Yes'})
day_df.yr = day_df.yr.replace({0: 2011, 1: 2012})
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah cuaca memengaruhi penggunaan sepeda sharing lebih besar pada pengguna biasa atau terdaftar?

```python
correlation = hour_df[['cnt', 'casual', 'registered', 'temp', 'atemp', 'hum', 'windspeed', 'weathersit']].corr()
print(correlation)

plt.scatter(hour_df['temp'], hour_df['cnt'])
plt.xlabel('Temperatur')
plt.ylabel('Jumlah Penyewaan')
plt.title('Korelasi Temperatur dan Jumlah Penyewaan')
plt.show()

plt.hist(hour_df['cnt'], bins=20)
plt.xlabel('Jumlah Penyewaan')
plt.ylabel('Frekuensi')
plt.title('Distribusi Jumlah Penyewaan Sepeda')
plt.show()

plt.figure(figsize=(8, 6))
plt.boxplot([hour_df[hour_df['casual'] > 0]['cnt'], hour_df[hour_df['registered'] > 0]['cnt']],
            labels=['Casual', 'Registered'], showmeans=True)
plt.ylabel('Jumlah Penyewaan')
plt.title('Perbandingan Distribusi Penyewaan antara Casual dan Registered')
plt.show()
```

### Pertanyaan 2: Bagaimana pola penggunaan sepeda sharing di Washington DC berubah setiap hari selama seminggu?

```python
if 'cnt' in day_df.columns:
  group_day = day_df.groupby('weekday')[['cnt', 'casual', 'registered']].mean()
  print(group_day)

  plt.plot(day_df.groupby('weekday')['cnt'].mean())
  plt.xlabel('Hari dalam Seminggu')
  plt.ylabel('Jumlah Penyewaan')
  plt.title('Tren Penggunaan Sepeda Per Hari dalam Seminggu')
  plt.show()

  plt.bar(group_day.index, group_day['cnt'])
  plt.xlabel('Hari dalam Seminggu')
  plt.ylabel('Jumlah Penyewaan')
  plt.title('Rata-Rata Penggunaan Sepeda Per Hari')
  plt.show()
```

**Insight:**
- Pada EDA dengan file hour_csv, dilakukan korelasi matriks, scatter plot, histogram serta Boxplot,   dimana menggunakan kolom cnt (total penyewaan sepeda), registered(pengguna/penyewa terdaftar), temp(suhu), casual(pengguna/penyewa biasa), dan hum(kelembaban). Pada korelasi matriks dan scatter plot menghasilkan bahwa ketika jumlah pengguna terdaftar meningkat, total penyewaan sepeda juga meningkat. Serta dikaitkan dengan suhu dan kelembaban, bahwa suhu yang hangat akan membuat jumlah penyewaan sepeda menjadi lebih tinggi dan sebaliknya. Jika kelembaban lebih tinggi, jumlah penyewaan sepeda menjadi lebih rendah.
Pada histogram juga menunjukkan ada beberapa hari dimana jumlah penyewaan sepeda melambung sangat tinggi daripada hari lainnya. Begitupun dengan Boxplot yang menghasilkan bahwa pengguna/penyewa terdaftar memiliki jumlah rata-rata penyewaan lebih tinggi dan lebih luas dibandingkan dengan pengguna/penyewa biasa.
- Dilakukan juga pada file day.csv untuk mendapatkan rata-rata penyewaan per hari dengan memasukkan jumlah rata-rata penyewaan sepeda (cnt) setiap hari dalam seminggu dan rata-rata penyewaan dari pengguna/penyewa biasa dan pengguna/penyewa terdaftar. Telah didapatkan bahwa rata-rata tertinggi terjadi pada hari Jumat dan Kamis, sedangkan Minggu dan Senin memiliki jumlah rata-rata penyewaan terendah.
